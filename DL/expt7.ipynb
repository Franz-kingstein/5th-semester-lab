```python
import gensim
from gensim.models import Word2Vec
from gensim.models import KeyedVectors
from gensim.test.utils import common_texts
```

```python
model_w2v = Word2Vec(sentences = common_texts, vector_size = 100, window = 5, min_count = 1, workers = 4)
model_w2v.save("First_vector_model.model")
```

```python
import numpy as np
embedding_index = {}
with open('/home/franz/Documents/Lab/DL/glove.6B.100d.txt', 'r', encoding='utf-8') as f:
     for line in f:
        values = line.rstrip().split()
        if not values:
            continue
        word = values[0]
        try:
            # pass the list of strings as a single argument to np.array
            embedding_index[word] = np.array(values[1:], dtype='float32')
        except ValueError:
            # skip malformed lines
            continue

# use get to avoid KeyError if the word is not present
print(embedding_index.get('computer'))
```
